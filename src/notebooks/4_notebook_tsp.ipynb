# 4 - Discrete: Traveling Salesman Algorithm Experiments
### Three test suites: Sparse · Clustered · Euclidean

```bash
python tsp_experiments.py                        # all cases, 100 iter, 60s timeout
python tsp_experiments.py --case sparse
python tsp_experiments.py --iter 200 --timeout 120
```

Result statuses stored per task:
| status | meaning |
|---|---|
| `ok` | finished, finite distance |
| `inf` | finished but returned inf/NaN |
| `timeout` | exceeded wall-clock limit |
| `error` | unhandled exception |
