# 🗺️ Travelling Salesman Problem — Algorithm Comparison

**Problem:** Find the shortest route visiting all cities exactly once and returning to the start. A classic NP-hard combinatorial optimisation problem.

**Properties:** Discrete · Combinatorial · Permutation search space

**Challenge:** Solution space grows factorially with city count. Tests how well algorithms handle discrete, large-scale search under time constraints.

**Test cases:** Sparse · Clustered · Euclidean

---
**Run before this notebook:**
```bash
python prerun_tsp_experiments.py                         # all cases, 100 iter, 60s timeout
python prerun_tsp_experiments.py --case sparse
python prerun_tsp_experiments.py --iter 200 --timeout 120
```
