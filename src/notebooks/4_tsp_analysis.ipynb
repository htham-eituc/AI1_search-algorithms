## Cross-case comparison

````python
import matplotlib.pyplot as plt
import pandas as pd

df = viz.df[viz.df['test_id'] != 'test_0'].dropna(subset=['best_fitness'])

from graph_visualize import ALGO_COLORS, ALGO_SHORT, ALGO_MARKERS

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, metric, ylabel in [
    (axes[0], 'best_fitness',   'Tour Distance'),
    (axes[1], 'execution_time', 'Time (s)'),
]:
    for case, grp in df.groupby('case'):
        pivot = grp.groupby('algorithm')[metric].mean().sort_values()
        xs = [ALGO_SHORT.get(a, a) for a in pivot.index]
        ls = '--' if case == 'clustered' else '-'
        for a, y in zip(pivot.index, pivot.values):
            ax.scatter(ALGO_SHORT.get(a,a), y,
                       color=ALGO_COLORS.get(a,'gray'), s=80,
                       marker=ALGO_MARKERS.get(a,'o'),
                       label=f'{case}-{ALGO_SHORT.get(a,a)}')
    ax.set_title(f'Mean {ylabel} — sparse vs Clustered')
    ax.set_ylabel(ylabel)
    ax.tick_params(axis='x', rotation=30)

from matplotlib.lines import Line2D
sparse_patch = Line2D([0],[0], linestyle='-',  color='gray', label='sparse')
clu_patch = Line2D([0],[0], linestyle='--', color='gray', label='Clustered')
axes[0].legend(handles=[sparse_patch, clu_patch], fontsize=9)

fig.suptitle('Cross-Case Comparison', fontsize=13, fontweight='bold')
fig.tight_layout()
fig.savefig('results/cross_case_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
````
